```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('Job_Placement_Data.csv')
```

```python
df
```

```python
df.isnull().sum()
```

### Handling Categorical Features

```python
gen = pd.get_dummies(df['gender'],drop_first = True)
hsc_sub = pd.get_dummies(df['hsc_subject'],drop_first=True)
ug = pd.get_dummies(df['undergrad_degree'],drop_first=True)
exp = pd.get_dummies(df['work_experience'],drop_first=True)
special = pd.get_dummies(df['specialisation'],drop_first=True)
```

```python
df = df.drop(['gender','hsc_subject','undergrad_degree','work_experience','ssc_board','hsc_board','specialisation'],axis=1)
```

```python
df
```

```python
df = pd.concat([gen,hsc_sub,ug,exp,special,df],axis=1)
df
```

### Splitting Dataset

```python
X = df.iloc[:,:-1].values
y = df.iloc[:,-1]
```

```python
y=pd.get_dummies(y,drop_first=True)
```

```python
y=y.values
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
```

## Grid Search CV

```python
It is for make combinatio
```

```python
from sklearn.model_selection import GridSearchCV
parameters = {'penalty':['l1','l2','elasticnet'],'C':[1,2,3,4,5,10,20,30,40,50,60,100],'max_iter':[100,200,300]}
classifier_reg = GridSearchCV(classifier,param_grid=parameters,scoring='accuracy',cv=5)
```

```python
classifier_reg.fit(X_train,y_train)
```

```python
print(classifier_reg.best_params_)
```

```python
print(classifier_reg.best_score_)
```

```python
y_pred = classifier_reg.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
score=accuracy_score(y_pred,y_test)
```

```python
print(score)
```

```python
print(classification_report(y_pred,y_test))
```
